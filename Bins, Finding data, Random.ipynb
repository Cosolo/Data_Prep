# Bining Data

```python
import pandas as pd
import numpy as np

data=pd.read_csv('Data/gradedata.csv')
```

```python
data.head(2)
```

Define BINS that we can from score get Grade!!

```python
#first let's see range of score grades
data.grade.unique()
```

```python
#creating bins
bins=[0,60,70,80,90,100]
#create names for groups
name_bins=['F','D','C','B','A']
```

```python
#NOW WE WILL ADD COLUMN TO DATA TABLE

data['letter_grade']=pd.cut(data['grade'], bins, labels=name_bins)
```

```python
data.head(2)
```

```python
pd.value_counts(data['letter_grade'])
```

```python
# create new column which says if student pass or failed exam. Score for pass is 80 or above.

bins2=[0,80,100]
bins2_name=['Fail','Pass']

data['Result_pass_or_fail']=pd.cut(data['grade'], bins2, labels=bins2_name)
```

```python
data.head(10)
```

Find average hours of study by letter grade!

```python
data.groupby('letter_grade')['hours'].mean()
```

```python
gender_preScore=data['grade'].groupby(data['grade'])
gender_preScore.mean()
```

Napravi novi stupac gdje ako je > 70 prolazi, ako je <=70 pada. Then compute mean hours of exercise of the female students with status of passing.

# Ranking Rows of Data

Create a column ranking by grade

```python
data['graderanked']=data['grade'].rank(ascending=1)
```

```python
data.tail()
```

```python
#bottom 20 students

data[data['graderanked'] < 21]
```

```python
#bottom 6 students in order

data[data['graderanked'] < 6].sort_values('graderanked')
```

```python
data['hours_ranked']=data['hours'].rank(ascending=1)
```

```python
data.head(10)
```

Create YES/NO is Failing Column

```python
data['IsFailing']=np.where(data['grade']<70, 'yes', 'no')
```

```python
data.head(5)
```

```python
#male students with failing scores
```

```python
data['IsFailingMale']=np.where((data['grade'] < 70) & (data['gender']=='male'), 'yes', 'no')
```

```python
data.head(3)
```

Create column timemgmt shows busy if student exercises more than 3 h per week and studies more than 17 h per week

```python
data['timemgmt']=np.where((data['exercise'] > 3) & (data['hours'] > 17), 'busy','no')
```

```python
data.head(2)
```

# Making new columns using functions

Make new column that contains full name of student

```python
def singlename(fn, ln):
    return fn + " " + ln
```

```python
data['fullname']=singlename(data['fname'],data['lname'])
```

```python
data.head(2)
```

Create column called total time that adds together hours of study per week and hours of exercises per week

```python
def sumhours(eh,sh):
    return eh+sh
```

```python
data['total_time']=sumhours(data['exercise'], data['hours'])
```
