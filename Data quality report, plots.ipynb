Find data type of each column

Count number of missing values

```python
data.isnull().sum()
```

```python
missing_data_counts=pd.DataFrame(data.isnull().sum(), columns=['Missing Values'])
missing_data_counts
```

Count number of present observations by column

```python
present_data_count=pd.DataFrame(data.count(), columns=['Present values'])
```

```python
# min value of each column
```

```python
min_values=pd.DataFrame(columns=['Min values'])
for v in list (data.columns.values):
    min_values.loc[v]=[data[v].min()]
min_values
```

```python
max_values=pd.DataFrame(columns=['Min values'])
for v in list (data.columns.values):
    max_values.loc[v]=[data[v].max()]
max_values
```

```python
#join all data into one table
```

```python
pd.concat([present_data_count,missing_data_counts,max_values,min_values], axis=1)
```
