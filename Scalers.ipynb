```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
uni=np.random.rand(100,2)
```

```python
plt.scatter(uni[:,0],uni[:,1])
```

```python
gaus=np.random.randn(100,2)
plt.scatter(gaus[:,0],gaus[:,1])
```

```python
from sklearn.preprocessing import Normalizer
```

```python
transformer = Normalizer().fit(gaus) # fit does nothing.
transformer
```

```python
gausN=transformer.transform(gaus)
plt.figure(figsize=[5,5])
plt.scatter(gausN[:,0],gausN[:,1])
```

```python
uniT = Normalizer().fit(uni)
uniN=uniT.transform(uni)
plt.figure(figsize=[5,5])
plt.scatter(uniN[:,0],uniN[:,1])
```
