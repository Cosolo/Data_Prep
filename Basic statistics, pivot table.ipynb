First Regression

```python
import statsmodels.formula.api as sm
```

```python
result=sm.ols(formula='grade~age + exercise + hours',data=data).fit()
result.summary()
```

Comment:

* R-squared - percentage of the variation in the data that can be accounted for by the regression (664 is good but not great)
* p-value (P > |t|) - represents the probability that the independet variable has no effect on the dependent variable - keep p- values under 5 %

```python
#rerun regression without age
```

```python
result2=sm.ols(formula='grade ~ exercise + hours', data=data).fit()
result2.summary()
```

* We haven't changed R squared, but we changed p-values. So, we can now look at our coefficients.
* grade=1.916 hours + .989 hours of exercise + 58.5316

```python
#min grade by gender
```

```python
pd.pivot_table(data, values=['grade'], index=['gender'], aggfunc='min')
```

```python
pd.pivot_table(data, index=['gender','age'], aggfunc='max', values=['hours'])
```
